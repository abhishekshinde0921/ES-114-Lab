Unifform PDF and CDF

```python
a=int(input())
b=int(input())
pdf=(1/((b-a)))
x=np.linspace(a,b,1000)
y=[]
for i in x:
    y.append(pdf)
plt.plot(x,y,color='b',label=f'Continuous_uniform ({a}, {b})')
plt.fill_between(x,y, alpha=0.1, color='b')
plt.xlabel("Random Variable (X)")
plt.ylabel("PDF Of Uniform random Variable")
plt.title("Probability Density Function (PDF) of Uniform Random Variable")
plt.legend()
plt.show()

z=(x-a)/(b-a)
y=y/np.sum(y)
cdf = np.cumsum(y)

plt.plot(x,z)
plt.fill_between(x,z, alpha=0.1, color='b',label='function')
plt.plot(x,cdf,'--')
plt.fill_between(x,cdf, alpha=0.1, color='r',label='cumsum')
plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Uniform random Variable")
plt.title("Cumulative Distribution Function (CDF) of Uniform Random Variable")
plt.grid(True)
plt.legend()
plt.show()



```

Gaussian PDF and CDF

```python
x=np.arange(-10,10.05,0.05)
y=[]
y1=[]
y2=[]

mew=float(input())
sigma=float(input())
mew1=float(input())
sigma1=float(input())
mew2=float(input())
sigma2=float(input())

for i in x:
    pmf=(math.e)**(-1/2*((i-mew)/sigma)**2)/((2*math.pi)**(1/2))
    y.append(pmf)
for i in x:
    pmf=(math.e)**(-1/2*((i-mew1)/sigma1)**2)/((2*math.pi)**(1/2))
    y1.append(pmf)
for i in x:
    pmf=(math.e)**(-1/2*((i-mew2)/sigma2)**2)/((2*math.pi)**(1/2))
    y2.append(pmf)

plt.plot(x,y,label=f'{mew}, {sigma}')
plt.fill_between(x,y, alpha=0.1, color='b')
plt.plot(x,y1,label=f'{mew1}, {sigma2}')
plt.fill_between(x,y1, alpha=0.1, color='r')
plt.plot(x,y2,label=f'{mew2}, {sigma2}')
plt.fill_between(x,y2, alpha=0.1, color='g')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Gaussian Random Variable')
plt.legend()
plt.grid(True)
plt.show()

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.plot(x,cdf)
plt.fill_between(x,cdf, alpha=0.1, color='b')

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Gaussian random Variable")
plt.title("Cumulative Distribution Function (CDF) of Gaussian Random Variable")
plt.grid(True)
plt.show()
```

Exponential PDF and CDF

```python
lamda=float(input())
lamda1=float(input())
lamda2=float(input())
x=np.linspace(0,10.05,1000)
y=[]
y1=[]
y2=[]

for i in x:
    pmf=lamda*(math.e)**(-lamda*i)
    y.append(pmf)
for i in x:
    pmf=lamda1*(math.e)**(-lamda1*i)
    y1.append(pmf)
for i in x:
    pmf=lamda2*(math.e)**(-lamda2*i)
    y2.append(pmf)
plt.plot(x,y,color="r",label=lamda)
plt.fill_between(x,y, alpha=0.1, color='r')
plt.plot(x,y1,color="g",label=lamda1)
plt.fill_between(x,y1, alpha=0.1, color='g')
plt.plot(x,y2,color="b",label=lamda2)
plt.fill_between(x,y2, alpha=0.1, color='b')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Exponential Random Variable')
plt.legend()
plt.grid(True)
plt.show()

z=[]
for i in x:
    pmf=1-(math.e)**(-lamda*i)
    z.append(pmf)

y=y/np.sum(y)
cdf = np.cumsum(y)

plt.plot(x,cdf,'--')
plt.fill_between(x,cdf, alpha=0.1, color='b',label='cumsum')
plt.plot(x,z)
plt.fill_between(x,z, alpha=0.1, color='r',label='function')

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Exponential random Variable")
plt.title("Cumulative Distribution Function (CDF) of Exponential Random Variable")
plt.grid(True)
plt.legend()
plt.show()
```

Cauchy PDF and CDF

```python
median=float(input())
gamma=float(input())
median1=float(input())
gamma1=float(input())
median2=float(input())
gamma2=float(input())

x=np.arange(-10,10.05,0.05)
y=[]
y1=[]
y2=[]
for i in x:
    pmf=1/((math.pi*gamma)*(1+((i-median)/gamma)**2))
    y.append(pmf)
for i in x:
    pmf=1/((math.pi*gamma1)*(1+((i-median1)/gamma1)**2))
    y1.append(pmf)
for i in x:
    pmf=1/((math.pi*gamma2)*(1+((i-median2)/gamma2)**2))
    y2.append(pmf)
plt.plot(x,y,color="r",label=f'{median}, {gamma}')
plt.fill_between(x,y, alpha=0.1,color='r')
plt.plot(x,y1,color="g",label=f'{median1}, {gamma2}')
plt.fill_between(x,y1, alpha=0.1, color='g')
plt.plot(x,y2,color="b",label=f'{median2}, {gamma2}')
plt.fill_between(x,y2, alpha=0.1, color='b')
plt.xlabel("Radnom Variable (X)")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Cauchy Random Variable")
plt.legend()
plt.grid(True)
plt.show()

z=[]
for i in x:
  u=0.5 + np.arctan((i-median)/gamma)/np.pi
  z.append(u)

y=y/np.sum(y)
cdf = np.cumsum(y)

plt.plot(x,cdf,'--')
plt.fill_between(x,cdf, alpha=0.1, color='b',label='cumsum')
plt.plot(x,z)
plt.fill_between(x,z, alpha=0.1, color='r',label='function')

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Cauchy random Variable")
plt.title("Cumulative Distribution Function (CDF) of Cauchy Random Variable")
plt.grid(True)
plt.legend()
plt.show()
```

Laplacian PDF and CDF

```python
u=int(input())
b=int(input())
u1=int(input())
b1=int(input())
u2=int(input())
b2=int(input())

x=np.arange(-11,11,0.05)
y=[]
y1=[]
y2=[]

for i in x:
    pmf=(math.e**((-(abs(i-u))/b)))/(2*b)
    y.append(pmf)
for i in x:
    pmf=(math.e**((-(abs(i-u1))/b1)))/(2*b1)
    y1.append(pmf)
for i in x:
    pmf=(math.e**((-(abs(i-u2))/b2)))/(2*b2)
    y2.append(pmf)

plt.plot(x,y,color="r",label=f'{b}, {u}')
plt.fill_between(x,y, alpha=0.1,color='r')
plt.plot(x,y1,color="g",label=f'{b1}, {u1}')
plt.fill_between(x,y1, alpha=0.1, color='g')
plt.plot(x,y2,color="b",label=f'{b2}, {u2}')
plt.fill_between(x,y2, alpha=0.1, color='b')
plt.xlabel("Radnom Variable (X)")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Laplacian Random Variable")
plt.legend()
plt.grid(True)
plt.show()

z=[]
for i in x:
  if i<u:
    z.append( 0.5*np.exp((i-u)/b))
  else:
    z.append(1-0.5*np.exp((u-i)/b))

y=y/np.sum(y)
cdf = np.cumsum(y)

plt.plot(x,cdf,'--')
plt.fill_between(x,cdf, alpha=0.1, color='b',label='cumsum')
plt.plot(x,z)
plt.fill_between(x,z, alpha=0.1, color='r',label='function')

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Laplacian random Variable")
plt.title("Cumulative Distribution Function (CDF) of Laplacian Random Variable")
plt.grid(True)
plt.legend()
plt.show()
```

Rayleigh PDF and CDF

```python
s=float(input())
s1=float(input())
s2=float(input())

x=np.arange(0,11,0.05)
y=[]
y1=[]
y2=[]

for i in x:
    pmf= (i*math.e**((-i**2)/(2*(s**2))))/(s**2)
    y.append(pmf)
for i in x:
    pmf= (i*math.e**((-i**2)/(2*(s1**2))))/(s1**2)
    y1.append(pmf)
for i in x:
    pmf= (i*math.e**((-i**2)/(2*(s2**2))))/(s2**2)
    y2.append(pmf)

plt.plot(x,y,color="r",label=f'{s}')
plt.fill_between(x,y, alpha=0.1,color='r')
plt.plot(x,y1,color="g",label=f'{s1}')
plt.fill_between(x,y1, alpha=0.1, color='g')
plt.plot(x,y2,color="b",label=f'{s2}')
plt.fill_between(x,y2, alpha=0.1, color='b')
plt.xlabel("Radnom Variable (X)")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Rayleigh Random Variable")
plt.legend()
plt.grid(True)
plt.show()

z=[]
for i in x:
  z.append(1-np.exp(-i*i/(2*s*s)))

y=y/np.sum(y)
cdf = np.cumsum(y)

plt.plot(x,cdf,'--')
plt.fill_between(x,cdf, alpha=0.1, color='b',label='cumsum')
plt.plot(x,z)
plt.fill_between(x,z, alpha=0.1, color='r',label='function')

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Rayleigh random Variable")
plt.title("Cumulative Distribution Function (CDF) of Rayleigh Random Variable")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Discrete Uniform CDF

```python
a=int(input())
b=int(input())

x=np.arange(a,b+1)
pmf=1/(b-a-1)
y=[]
for i in x:
  y.append(pmf)

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Discrete Uniform Variable")
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Variable")
plt.grid(True)
plt.show()
```

Bernoulli CDF

```python
p=float(input())

x=np.array([-2,-1,0,1,2,3])
y=np.array([0,0,0,p,1-p,0])

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)
plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Bernoulli Variable")
plt.title("Cumulative Distribution Function (CDF) of Bernoulli Variable")
plt.grid(True)
plt.show()
```

Binomial CDF

```python
n=int(input())
p=float(input())

x=np.arange(0,n+1)

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def Binomial_distribution(n,p):
  r=[]
  for i in range(0,n+1):
    x=(factorial(n)/(factorial(n-i)*factorial(i)))*(p**i)*((1-p)**(n-i))
    r.append(x)
  return(r)

y=Binomial_distribution(n,p)

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Binomial Variable")
plt.title("Cumulative Distribution Function (CDF) of Binomial Variable")
plt.grid(True)
plt.show()
```

Poission CDF

```python
n=int(input())
p=float(input())
n1=float(n)

x=np.arange(0,n+1)

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def poisson(n,p):
    lamda=n1*p
    r=[]
    for i in range(0,n+1):
      z=(np.exp(-lamda))*(lamda**i)/(factorial(i))
      r.append(z)
    return(r)

y=poisson(n,p)

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Poission Variable")
plt.title("Cumulative Distribution Function (CDF) of Poission Variable")
plt.grid(True)
plt.show()
```

Geometric CDF

```python
n=int(input())
p=float(input())

x=list(range(1,n+1))

def geometric(n,p):
  r=[]
  for i in range(1,n+1):
    z=p*((1-p)**(i-1))
    r.append(z)
  return(r)

y=geometric(n,p)

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)

plt.xlabel("Random Variable (X)")
plt.ylabel("CDF Of Geometric Variable")
plt.title("Cumulative Distribution Function (CDF) of Geometric Variable")
plt.grid(True)
plt.show()
```
