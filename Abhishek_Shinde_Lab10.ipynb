**Q3**

Plot a graph of number of matches played per tournament in the year 2010.

```python
df = pd.read_excel("atp_2010.xls")
matches_per_tournament = df.groupby('Tournament')['Winner'].count().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(matches_per_tournament['Tournament'], matches_per_tournament['Winner'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Tournament')
plt.ylabel('Number of Matches')
plt.title('Number of Matches per Tournament')
plt.tight_layout()
plt.show()
```

```python
# @title
main_data = []
for year in range(2000, 2013):
    filename = f"atp_{year}.xls"
    df = pd.read_excel(filename)
    main_data.append(df)
for year in range(2013,2025):
    filename = f"atp_{year}.xlsx"
    df = pd.read_excel(filename)
    main_data.append(df)
df_players=pd.read_csv("atp_players.csv")
final_merge=pd.concat(main_data)
```

**Q4**

Plot the types of rounds played by players from year 2000 to 2024.

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='Round', data=final_merge, order=final_merge['Round'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Round')
plt.title('Distribution of Rounds Reached by Players')
plt.show()
```
