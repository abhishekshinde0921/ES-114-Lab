Discrete Uniform PMF

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

a = 1
b= 6

def discrete_uniform_pmf( a, b):
  return 1 / (b - a + 1)

w=list(range(a,b+1))
l=[]
for i in range(len(w)):
  l.append(discrete_uniform_pmf(a,b))


def plot_discrete_uniform(w,l):

  plt.bar(w,l,label=f'Descrete_uniform ({a}, {b})')

  plt.title('Probability Mass Function of Discrete Uniform Random Variable')
  plt.legend()
  plt.xlabel('Outcome (x)')
  plt.ylabel('Probability')

  plt.grid(True)
  plt.show(True)

plot_discrete_uniform(w,l)
```

Binomial PMF

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

n=10
p=0.5;
x=list(range(0,n+1))

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def Binomial_distribution(n,p):
  r=[]
  for i in range(0,n+1):
    x=(factorial(n)/(factorial(n-i)*factorial(i)))*(p**i)*((1-p)**(n-i))
    r.append(x)
  return(r)

y=Binomial_distribution(n,p)

def plot_Binomial(x,y):
    plt.bar(x,y,width=0.2,label=f'Binomial ({0}, {n})')
    plt.title('Probability Mass Function of Binomial Variable')
    plt.xlabel('Outcome (x)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show(True)

plot_Binomial(x,y)
```

Poisson PMF

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

n=140
p=0.5
n1=float(n)

x=list(range(0,n+1))

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def poisson(n,p):
    lamda=n1*p
    r=[]
    for i in range(0,n+1):
      z=(np.exp(-lamda))*(lamda**i)/(factorial(i))
      r.append(z)
    return(r)

y=poisson(n,p)

def plot_poisson(x,y):
    plt.bar(x,y,width=0.2,label=f'poisson ({0}, {n})')
    plt.title('Probability Mass Function of Poisson Variable')
    plt.xlabel('Outcome (x)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show(True)
plot_poisson(x,y)
```

Geometric PMF

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

n=10
p=0.4

x=list(range(1,n+1))

def geometric(n,p):
  r=[]
  for i in range(1,n+1):
    z=p*((1-p)**(i-1))
    r.append(z)
  return(r)

y=geometric(n,p)


def plot_geometric(x,y):
    plt.bar(x,y,width=0.2,label=f'geometric ({1}, {n})')
    plt.title('Probability Mass Function of Geometric Variable')
    plt.xlabel('Outcome (x)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show(True)

plot_geometric(x,y)
```

Equivalence of Binomial and Poisson

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
n=140
p=0.4
n1=float(n)

xb=list(range(0,n+1))

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def Binomial_distribution(n,p):
  r=[]
  for i in range(0,n+1):
    x=(factorial(n1)/(factorial(n1-i)*factorial(i)))*(p**i)*((1-p)**(n1-i))
    r.append(x)
  return(r)

yb=np.array(Binomial_distribution(n,p))

def plot_Binomial(x,y):
    plt.bar(x,y,width=0.2,label=f'Binomial ({0}, {n})')
    plt.title('Probability Mass Function of Binomial Variable')
    plt.xlabel('Outcome (x)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show(True)

plot_Binomial(xb,yb)

xp=list(range(0,n+1))

def factorial(n):
  if n ==0:return 1
  return n*factorial(n-1)

def poisson(n,p):
    lamda=n1*p
    r=[]
    for i in range(0,n+1):
      z=(np.exp(-lamda))*(lamda**i)/(factorial(i))
      r.append(z)
    return(r)

yp=np.array(poisson(n,p))

def plot_poisson(x,y):
    plt.bar(x,y,width=0.2,label=f'poisson ({0}, {n})')
    plt.title('Probability Mass Function of Poisson Variable')
    plt.xlabel('Outcome (x)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)
    plt.show(True)
plot_poisson(xp,yp)

print(np.sqrt(np.average((yb-yp)**2)),"is the RMS value")
```
